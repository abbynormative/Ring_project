```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("Ring San Francisco Log Sheet (Responses).csv")
df.head()
```

```python
df.rename(index=str, columns={"What is the title of this video post?": "Post title"}, inplace=True)
df.rename(index=str, columns={"Subjects (check as many as applicable)": "Subjects"}, inplace=True)
df.rename(index=str, columns={"What is the date of this video post": "Post date"}, inplace=True)
df.rename(index=str, columns={"Please describe the inaccuracy if one was observed": "Observed inaccuracies"}, inplace=True)
```

```python
df.head()
```

```python
ax=sns.countplot(data=df,x='Visibility')
#plt.savefig(".png")
ax.set_xticklabels(ax.set_xticklabels(['poorly lit','clearly lit','partially lit','badly lit']),rotation=90)
plt.show()
#combine poorly and badly lit
```

```python
ax=sns.countplot(data=df,x='Content')
#ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
ax.set_xticklabels(ax.set_xticklabels(['possible crime','no evidence of crime','shows a crime']),rotation=90)
plt.show()
```
